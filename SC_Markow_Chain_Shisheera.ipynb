```r
library(readxl)

xls_file <- "rspb20141370supp2.xls"

# All species
species_list <- c("Bat", "English", "Finch",
                  "Chickadee", "Pilot",
                  "Orangutan", "Orca", "Hyrax")

# Store results
results <- list()

for (sp in species_list) {

  cat("\n", rep("=", 40), "\n", sep="")
  cat("SPECIES:", sp, "\n")
  cat(rep("=", 40), "\n", sep="")

  # Load
  raw <- read_excel(xls_file,
                    sheet     = sp,
                    col_names = FALSE)

  seq <- unlist(raw)
  seq <- seq[!is.na(seq)]
  seq <- as.character(as.integer(seq))

  cat("Total elements:", length(seq), "\n")
  cat("Unique states: ",
      length(unique(seq)), "\n\n")

  # Transition matrix
  states    <- sort(unique(seq))
  n         <- length(states)
  count_mat <- matrix(0, nrow=n, ncol=n,
                      dimnames=list(states,states))

  for (i in 1:(length(seq)-1)) {
    count_mat[seq[i], seq[i+1]] <-
      count_mat[seq[i], seq[i+1]] + 1
  }

  prob_mat <- round(
    count_mat / rowSums(count_mat), 3
  )

  # Self transitions
  mean_self <- round(mean(diag(prob_mat)), 3)

  cat("Transition Probability Matrix:\n")
  print(prob_mat)

  cat("\nMean self-transition:", mean_self, "\n")

  if (mean_self > 0.30) {
    cat("→ HIGH repetition — supports RP\n")
  } else {
    cat("→ LOW repetition — supports Markov\n")
  }

  # Save
  results[[sp]] <- list(
    prob_mat  = prob_mat,
    mean_self = mean_self,
    n_elements = length(seq),
    n_states   = length(states)
  )
}

#SUMMARY TABLE
cat("\n\n", rep("=", 40), "\n", sep="")
cat("SUMMARY — ALL SPECIES\n")
cat(rep("=", 40), "\n\n", sep="")

cat(sprintf("%-12s %-10s %-10s %-10s %-15s\n",
    "Species", "Elements", "States",
    "Mean Self", "Interpretation"))
cat(rep("-", 60), "\n", sep="")

for (sp in species_list) {
  r <- results[[sp]]
  interp <- ifelse(r$mean_self > 0.30,
                   "Supports RP",
                   "Supports Markov")
  cat(sprintf("%-12s %-10d %-10d %-10.3f %-15s\n",
      sp,
      r$n_elements,
      r$n_states,
      r$mean_self,
      interp))
}
```
